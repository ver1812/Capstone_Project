# File to download All data Locally.

```python
import os
from datasets import load_dataset
import pandas as pd

RAW_DIR = "raw"
os.makedirs(RAW_DIR, exist_ok=True)
```

# Deepset Dataset

```python
deepset = load_dataset("deepset/prompt-injections")
print(deepset)
```

```python
deepset_dfs = []
for split_name in deepset.keys():
    split_df = deepset[split_name].to_pandas()
    split_df["split"] = split_name
    deepset_dfs.append(split_df)

deepset_df = pd.concat(deepset_dfs, ignore_index=True)
print(deepset_df.shape)
deepset_df.head()
```

```python
deepset_df.to_csv(f"{RAW_DIR}/deepset_prompt_injections.csv", index=False)
print("Saved deepset_prompt_injections.csv")
```
