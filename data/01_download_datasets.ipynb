# File to download All data Locally.

# Most of the code used here is courtesy of Huggingface sample code

```python
import os
from datasets import load_dataset
import pandas as pd

RAW_DIR = "raw"
os.makedirs(RAW_DIR, exist_ok=True)
```

# Deepset Dataset

```python
deepset = load_dataset("deepset/prompt-injections")
print(deepset)
```

```python
deepset_dfs = []
for split_name in deepset.keys():
    split_df = deepset[split_name].to_pandas()
    split_df["split"] = split_name
    deepset_dfs.append(split_df)

deepset_df = pd.concat(deepset_dfs, ignore_index=True)
print(deepset_df.shape)
deepset_df.head()
```

```python
deepset_df.to_csv(f"{RAW_DIR}/deepset_prompt_injections.csv", index=False)
print("Saved deepset_prompt_injections.csv")
```

# Wildjailbreak

You need active login token to access.

```python
wildjailbreak = load_dataset("allenai/wildjailbreak", "train", delimiter="\t", keep_default_na=False)
print(wildjailbreak)
```

```python
wjb_df = wildjailbreak["train"].to_pandas()
print(wjb_df.shape)
wjb_df.head()
```

```python
wjb_df.to_csv(f"{RAW_DIR}/wildjailbreak_raw.csv", index=False)
print("Saved wildjailbreak_raw.csv")
```

```python
malicious_prompts = load_dataset("ahsanayub/malicious-prompts")
print(malicious_prompts)
```

```python
mp_train_df = malicious_prompts["train"].to_pandas()
mp_test_df = malicious_prompts["test"].to_pandas()

print("train:", mp_train_df.shape)
print("test:", mp_test_df.shape)
mp_train_df.head()
```

```python
mp_train_df.to_csv(f"{RAW_DIR}/malicious_prompts_train.csv", index=False)
mp_test_df.to_csv(f"{RAW_DIR}/malicious_prompts_test.csv", index=False)
print("Saved malicious_prompts_train.csv and malicious_prompts_test.csv")
```

```python
print("Files saved in data/raw/:")
for f in os.listdir(RAW_DIR):
    path = os.path.join(RAW_DIR, f)
    size_mb = os.path.getsize(path) / (1024 * 1024)
    print(f"  {f} — {size_mb:.1f} MB")
```
